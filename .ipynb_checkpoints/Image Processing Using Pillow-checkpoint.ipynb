```python
from PIL import ImageFont
from PIL import ImageDraw

watermark_image = image1.copy()
draw = ImageDraw.Draw(watermark_image)
# font = ImageFont.trutype('arial.ttf', 100) # select font
font = ImageFont.truetype("/usr/share/fonts/truetype/freefont/FreeMono.ttf", 100, encoding="unic")

# draw.text((0, 0), "Amir Savvy", (0,0,0), font=font)
draw.text((0,0), u'Hello World!', 'white', font)

plt.imshow(watermark_image)
```

```python
size = (250, 250)
croped_image = image2.copy()
croped_image.thumbnail(size)

copied_image = image1.copy()
copied_image.paste(croped_image, (0, 0))
plt.imshow(copied_image)
```
